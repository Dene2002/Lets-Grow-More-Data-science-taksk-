```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
import cv2
```

```python
image = cv2.imread('D:\Internships and job\Lets Grow More\Task 2/Image.jpg')
#cv2.imshow("Scenary", image)
```

```python
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.axis('off')
plt.show()
```

```python
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
plt.imshow(cv2.cvtColor(gray_image, cv2.COLOR_BGR2RGB))
plt.axis('off')
plt.show()
```

```python
inversion_image = 255 - gray_image
plt.imshow(cv2.cvtColor(inversion_image, cv2.COLOR_BGR2RGB))
plt.axis('off')
plt.show()
```

```python
blurr = cv2.GaussianBlur(inversion_image, (21,21), 0)

#Invert the blurred image
inversion_blur = 255 - blurr
final_image = cv2.divide(gray_image, inversion_blur, scale=256.0)
plt.imshow(cv2.cvtColor(final_image, cv2.COLOR_BGR2RGB))
plt.axis('off')
plt.show()
```

```python
cv2.imshow("original image", image)
cv2.imshow("pencil sketch", final_image)
```
